```python
import pandas as pd
df = pd.read_excel("bitcoin.xlsx")
df.rename(columns={
    "timeOpen": "timestamp",
    "priceOpen": "open",
    "priceHigh": "high",
    "priceLow": "low",
    "priceClose": "close"
}, inplace=True)

time_cols = ["timestamp", "timeClose", "timeHigh", "timeLow"]
for col in time_cols:
    df[col] = pd.to_datetime(df[col], unit="ms")

cols = ["open", "high", "low", "close"]
df[cols] = df[cols].astype(float)
df.sort_values("timestamp", inplace=True)


df["SMA_20"] = df["close"].rolling(window=20).mean() #Simple Moving Average (20 days)
df["SMA_50"] = df["close"].rolling(window=50).mean() #Simple Moving Average (50 days)

df["signal"] = 0
df.loc[df["SMA_20"] > df["SMA_50"], "signal"] = 1 #купити
df.loc[df["SMA_20"] < df["SMA_50"], "signal"] = -1 #продати
df["position"] = df["signal"].shift(1)

df["price_change"] = df["close"].diff()
df["profit"] = df["price_change"] * df["position"]

df["overoll_profit"] = df["profit"].cumsum()
df["overoll_price_change"] = df["price_change"].cumsum()

display(df.tail(10))
```
